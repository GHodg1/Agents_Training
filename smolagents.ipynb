## 2. Creating an agent using smolagents

The code for the Agent is stored in the First_agent_template directory.

Here is a breakdown of the file structure:

- SYSTEM PROMPT: stored in prompts.yaml - contains predefined instructions on agent behaviour. Using YAML files allows for easy customisation and re-use

- app.py: Stores the setup code for querying the LLM (Qwen/Qwen2.5-Coder-32B-Instruct), place for adding tools to the agent setup code.


```python
# Set up import path to the project root and load the agent
from pathlib import Path
import sys, os

# Ensure the project root is on sys.path
project_root = Path.cwd()
sys.path.insert(0, str(project_root))

# Map AGENT_API_TOKEN to HF_TOKEN if needed for Hub auth (no secret printed)
if not (os.getenv("HF_TOKEN") or os.getenv("HUGGING_FACE_HUB_TOKEN")):
    if os.getenv("AGENT_API_TOKEN"):
        os.environ["HF_TOKEN"] = os.getenv("AGENT_API_TOKEN")

from First_agent_template.app import agent

# Define the user query
user_query = "What is the current time in Tokyo?"
# Run the agent with the user query
response = agent.run(user_query)
print(response)
```
